```python
#2 write a program to reverse an array in place.in place means you cannot create a new arrey.you have to update the original array 

def reverseList(A, start, end):
    while start < end:
        A[start], A[end] = A[end], A[start]
        start += 1
        end -= 1

A = [1, 2, 3, 4, 5, 6]
print(A)
reverseList(A, 0, 5)
print("Reversed list is")
print(A)
```

```python
#8 write  a program to check if all the brackets are closed in a given code snippet.

def areBracketsBalanced(expr):
    stack = []
 
    for char in expr:
        if char in ["(", "{", "["]:
           stack.append(char)
        else:
            if not stack:
                return False
            current_char = stack.pop()
            if current_char == '(':
                if char != ")":
                    return False    
            if current_char == '{':
                if char != "}":
                    return False
            if current_char == '[':
                if char != "]":
                    return False
 
    if stack:
        return False
    return True
 
if _name_ == "_main_":
    expr = "{()}[]"
 
    if areBracketsBalanced(expr):
        print("Balanced")
    else:
        print("Not Balanced")
        
```

```python
#9 write  a program to reverse a stack.

class Stack:
 
    def _init_(self):
        self.Elements = []
         
    def push(self, value):
        self.Elements.append(value)
       
    def pop(self):
        return self.Elements.pop()
     
    def empty(self):
        return self.Elements == []
     
    def show(self):
        for value in reversed(self.Elements):
            print(value)
 

def BottomInsert(s, value):
   
    if s.empty():
       s.push(value)
     else:
        popped = s.pop()
        BottomInsert(s, value)
        s.push(popped)
 
def Reverse(s):
    if s.empty():
        pass
    else:
        popped = s.pop()
        Reverse(s)
        BottomInsert(s, popped)
 
stk = Stack()
 
stk.push(1)
stk.push(2)
stk.push(3)
stk.push(4)
stk.push(5)
 
print("Original Stack")
stk.show()
 
print("\nStack after Reversing")
Reverse(stk)
stk.show()     
```

```python
#10 write  a program to find the smallest number using a stack.

def find_min(stack): 
    if not stack:
        return None

min_val = stack.pop() 
while stack:
    val = stack.pop()
    if val < min_val:
       min_val = val

return min_val

```
